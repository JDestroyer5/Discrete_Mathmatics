a) I created a Dino Guesser. You choose one of five dinosuars, (Pteranodon, T-Rex, Spinosuarus, Triceratops, Stegosuarus) you follow the path to determine what dinosuar it is.

### Two's Complement

Using two's complement to subtract 25-37

First for checking later do the subtraction in decimal first.

```python
25-37
```

Convert both numbers to binary

```python
bin(25)
```

```python
bin(37)
```

Using two's complement convert 100101 to negative by fliping all bits and adding 1

```python
binB = bin(37)
flipbit = ''
for bit in binB:
  if bit == "1":
    flipbit += "0"
  else:
    flipbit += "1"
print (flipbit)
```

```python
11011010+1
```

Now we add them together to get the answer

```python
binsum=0b11011011+0b00011001
bin(binsum)
```

Since the first bit is a 1 the answer is a negative and you need to convert it back to positve by agian using two' complement.

```python
binB = bin(binsum)
flipbit = ''
for bit in binB:
  if bit == "1":
    flipbit += "0"
  else:
    flipbit += "1"
print (flipbit)
```

You only need the last 8 bits

```python
final=0b00001011+0b1
bin(final)
```

Finally we convert back to decimal and put a negative on it because of the step before.
